Markers: Representation of point in graph/chart

marker = 's' # s is for square

marker = '.' # . is for point

marker = '*' # * is for star

marker = 'o' # o is for circle

ms = marker size

mec = marker edge color

mfc = marker face color **bold text**

```python
import matplotlib.pyplot as plt

# Define data points for the x-axis and y-axis
x = [1,2,3,4,5,6]
y  =[12,19,5,16,25,13]

# Add labels and title
plt.xlabel('Power Play')
plt.ylabel('Runs')
plt.title('T20 match')

# Plot the data points specified by the arrays x and y, with markers represented as dots ('.')
# Set the size of the markers to 10 points (ms = 10)
# Set the edge color of the markers to red ('r') and the face color to black ('k')
plt.plot(x,y, marker = 'o', ms = 8 , mec = 'r', mfc = 'k')

# Display the plot
plt.show()
```

Lines:

ls = line style e.g. ls = "dashed" ls = "dotted" ls = "dashdot"

lw = line width c = color of the line

```python
import matplotlib.pyplot as plt

# Define data points for the x-axis and y-axis
x = [1,2,3,4,5,6]
y  =[12,19,5,16,25,13]

# Add labels and title
plt.xlabel('Power Play')
plt.ylabel('Runs')
plt.title('T20 match')

# Plot data points with square markers, red edges, black faces, dashed lines, and black color

plt.plot(x,y,marker = "s", ms= 10, mec = 'r', mfc = 'k', ls = "dashed", c = 'k', lw = 2)

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6]
y  =[12,19,5,16,25,13]

font_properties = {
    'family': 'serif',  # Font family ('serif', 'sans-serif', 'monospace', etc.)
    'color': 'blue',    # Text color
    'weight': 'bold',   # Font weight ('normal', 'bold', 'light', 'heavy')
    'size': 14,         # Font size
}

# Set font properties for x-axis label
plt.xlabel('Power Play', fontdict = font_properties)

# Set font properties for y-axis label
plt.ylabel('Runs',  fontdict = font_properties)

plt.plot(x,y,marker = "s", ms= 10, mec = 'r', mfc = 'k', ls = "dashed", c = 'k', lw = 2)

plt.show()
```

Grid function is used to get grid lines on the chart.

grid(axis = 'y')

grid(axis = 'x')

In the grid we can use the line properties such as lw= line width, c = color, ls = line style

```python
import matplotlib.pyplot as plt

# Data points
x = [1, 2, 3, 4, 5, 6]
y = [12, 19, 5, 16, 25, 13]

# Adding labels
plt.xlabel('Power Play')  # Label for x-axis
plt.ylabel('Runs')  # Label for y-axis

# Plotting data points with dots as markers, red edges, and black faces
plt.plot(x, y, marker='.', ms=10, mec='r', mfc='k')

# Displaying gridlines only along the y-axis
plt.grid(axis='y')

# Showing the plot
plt.show()
```
